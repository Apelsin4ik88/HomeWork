```python
import math

def calculate_series_sum(epsilon=1e-3):
    total_sum = 0.0
    n = 1
    while True:
        numerator = math.factorial(n)
        denominator = math.factorial(2 ** n)
        term = numerator / denominator
        total_sum += term
        print(f"n = {n}, term = {term}, sum = {total_sum}")
        if term < epsilon:
            break
        n += 1
    return total_sum
result = calculate_series_sum()
print(f"Сумма ряда с точностью 10^(-3): {result:.6f}")
```

```python
sum_positive_even = 0

while True:
    num = int(input("Введите число (0 для завершения): "))
    if num == 0:
        break
    if num > 0 and num % 2 == 0:
        sum_positive_even += num

print("Сумма положительных четных чисел:", sum_positive_even)
```
