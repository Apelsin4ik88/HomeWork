```python
'''Proc ...Сибай'''
def find_numbers():
    for num in range(1000, 10000):
        a = num // 1000
        b = (num // 100) % 10
        c = (num // 10) % 10
        d = num % 10
        if len({a, b, c, d}) == 4:
            if (a * 10 + b) - (c * 10 + d) == a + b + c + d:
                print(num)
```

```python
import math

def divide_fractions(A, B, C, D):
    numerator = A * D
    denominator = B * C
    gcd_value = math.gcd(numerator, denominator)
    return numerator // gcd_value, denominator // gcd_value
```

```python
def factorial_sum(n):
    total = 0
    factorial = 1
    for i in range(1, n + 1):
        factorial *= i
        total += factorial
    return total
```

```python
def PowerA3(A):
    B = A ** 3
    return B

numbers = [2.5, 3.0, 1.2, 4.1, 5.6]
for num in numbers:
    print(PowerA3(num))
```

```python
def PowerA234(A):
    B = A ** 2
    C = A ** 3
    D = A ** 4
    return B, C, D

numbers = [2.5, 3.0, 1.2, 4.1, 5.6]
for num in numbers:
    b, c, d = PowerA234(num)
    print(b, c, d)
```

```python
import math

def Mean(X, Y):
    AMean = (X + Y) / 2
    GMean = math.sqrt(X * Y)
    return AMean, GMean

A, B, C, D = 4.0, 9.0, 16.0, 25.0
print(Mean(A, B))
print(Mean(A, C))
print(Mean(A, D))
```

```python
import math

def TrianglePS(a):
    P = 3 * a
    S = (a ** 2) * math.sqrt(3) / 4
    return P, S

sides = [3.0, 5.0, 7.0]
for side in sides:
    perimeter, area = TrianglePS(side)
    print(perimeter, area)
```

```python
def RectPS(x1, y1, x2, y2):
    a = abs(x2 - x1)
    b = abs(y2 - y1)
    P = 2 * (a + b)
    S = a * b
    return P, S

rectangles = [
    (1.0, 1.0, 4.0, 5.0),
    (0.0, 0.0, 3.0, 4.0),
    (2.0, 3.0, 6.0, 8.0)
]
for rect in rectangles:
    perimeter, area = RectPS(*rect)
    print(perimeter, area)
```
